```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Path to your CSV file
# hem_path = '/Volumes/Siren/Brain_Analyses/Pouya_is_LITT/N235_S12_Large_SOX2/Cell-Centroids-Negative.csv'
# cell_path = '/Volumes/Siren/Brain_Analyses/Pouya_is_LITT/N235_S12_Large_SOX2/Cell-Centroids-Positive.csv'
hem_path = '/Volumes/Siren/Brain_Analyses/Pouya_is_LITT/N250_S13_Large_SOX2/Cell-Centroids-Negative.csv'
cell_path = '/Volumes/Siren/Brain_Analyses/Pouya_is_LITT/N250_S13_Large_SOX2/Cell-Centroids-Positive.csv'

# Read the CSV
hem_cells = np.genfromtxt(hem_path, delimiter=',', skip_header=1, dtype=[('CellID', 'int32'), ('X', 'float32'), ('Y', 'float32')])
dab_cells = np.genfromtxt(cell_path, delimiter=',', skip_header=1, dtype=[('CellID', 'int32'), ('X', 'float32'), ('Y', 'float32')])
```

```python
plt.figure()

# Plot HEM cells heat map in blue
sns.kdeplot(
    x=hem_cells['X'], y=hem_cells['Y'],
    cmap="Blues", fill=True, alpha=0.5, thresh=0.05, label="HEM"
)

# Plot DAB cells heat map in red
sns.kdeplot(
    x=dab_cells['X'], y=dab_cells['Y'],
    cmap="Reds", fill=True, alpha=0.5, thresh=0.05, label="DAB"
)

plt.legend()
plt.title("Density Heatmap: HEM Cells vs DAB Cells")

plt.gca().set_aspect('auto')  # Make axes proportional to data

plt.show()
```
